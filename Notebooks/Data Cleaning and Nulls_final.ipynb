```python
# Import necessary libraries
import pandas as pd

# Load the CSV file into a pandas DataFrame
file_path = "C:/Users/lenovo/Downloads/transcript_data-20240917T043654Z-001/transcript_data/7.csv"  # Replace with your file path
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to get an initial view of the data
print("First 5 rows of the data:")
print(df.head())

# Check for null values in each column
print("\nNull values in each column:")
null_values = df.isnull().sum()
print(null_values)

# Identify columns with null values
print("\nColumns with null values:")
null_columns = df.columns[df.isnull().any()].tolist()
print(null_columns)

# Get a summary of the DataFrame, including count, mean, standard deviation, min, max, etc.
print("\nSummary of the dataset (for numeric columns):")
summary_stats = df.describe()
print(summary_stats)

# Get a summary of categorical columns if there are any
print("\nSummary of the dataset (for categorical columns):")
categorical_summary = df.describe(include=['object'])
print(categorical_summary)

# Checking for any completely null columns (columns where all values are NaN)
print("\nCompletely null columns:")
completely_null_columns = df.columns[df.isnull().sum() == len(df)].tolist()
print(completely_null_columns)

# Checking for any duplicate rows
print("\nNumber of duplicate rows in the dataset:")
duplicate_rows = df.duplicated().sum()
print(duplicate_rows)

# Adding a new column 'net_score' by applying the formula
df['net_score'] = (df['positive'] + df['neutral'] + df['confident'] + 
                   df['concise'] + df['enthusiastic']) - (df['negative'] + df['hesitant'])

# Display the first few rows to confirm the new column is added
print("\nFirst 5 rows with the new 'net_score' column:")
print(df[['text', 'positive', 'neutral', 'confident', 'concise', 'enthusiastic', 'negative', 'hesitant', 'net_score']].head())

# Save the updated DataFrame to a new CSV file
output_file_path = "C:/Users/lenovo/Downloads/transcript_data-20240917T043654Z-001/transcript_data/7_new.csv"  # Replace with your desired output path
df.to_csv(output_file_path, index=False)

print(f"\nThe updated CSV file with the 'net_score' column has been saved to: {output_file_path}")
```
